```python
import random

class Account_KS:
    # (Q2) 클래스 변수: 생성된 계좌 개수
    account_count = 0

    def __init__(self, owner, balance):
        # (Q1) 생성자: 예금주와 초기 잔액만 입력
        self.bank = "SC은행"
        self.owner = owner
        self.balance = balance

        # 계좌번호 생성
        self.account_number = self._make_account_number()

        # (Q7) 입금 횟수
        self.deposit_count = 0

        # (Q10) 입금/출금 내역 저장 (각 항목: 딕셔너리로 저장)
        self.deposit_logs = []
        self.withdraw_logs = []

        # 계좌 생성 개수 증가
        Account_KS.account_count += 1

    # (Q1) 계좌번호 생성 메서드 (내부용)
    def _make_account_number(self):
        part1 = str(random.randint(0, 999)).zfill(3)
        part2 = str(random.randint(0, 99)).zfill(2)
        part3 = str(random.randint(0, 999999)).zfill(6)
        return f"{part1}-{part2}-{part3}"

    # (Q3) 클래스 메서드: 계좌 개수 출력
    @classmethod
    def get_account_num(cls):
        print(f"현재 생성된 계좌 개수: {cls.account_count}개")

    # (Q4) 입금
    def deposit(self, amount):
        if amount < 1:
            print("입금은 최소 1원 이상만 가능합니다.")
            return

        # 입금 처리
        self.balance += amount
        self.deposit_count += 1

        # (Q10) 내역 기록
        self.deposit_logs.append({
            "type": "입금",
            "amount": amount,
            "balance_after": self.balance
        })

        # (Q7) 입금 5회마다 이자 1% 지급
        if self.deposit_count % 5 == 0:
            interest = int(self.balance * 0.01)
            self.balance += interest

            # 이자도 내역으로 남기기 
            self.deposit_logs.append({
                "type": "이자",
                "amount": interest,
                "balance_after": self.balance
            })
            print(f"입금 5회 달성! 이자 {interest:,}원이 지급되었습니다.")

        print(f"{self.owner} 님 {amount:,}원 입금 완료 (잔액: {self.balance:,}원)")

    # (Q5) 출금
    def withdraw(self, amount):
        if amount < 1:
            print("출금은 최소 1원 이상만 가능합니다.")
            return

        if amount > self.balance:
            print("잔액이 부족하여 출금할 수 없습니다.")
            return

        # 출금 처리
        self.balance -= amount

        # (Q10) 내역 기록
        self.withdraw_logs.append({
            "type": "출금",
            "amount": amount,
            "balance_after": self.balance
        })

        print(f"{amount:,}원 출금 완료 (잔액: {self.balance:,}원)")

    # (Q6) 정보 출력
    def display_info(self):
        print(f"(은행이름: {self.bank}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔액: {self.balance:,}원)")

    # (Q10) 입금 내역 출력
    def deposit_history(self):
        print("===== 입금/이자 내역 =====")
        if len(self.deposit_logs) == 0:
            print("내역이 없습니다.")
            return

        for i, log in enumerate(self.deposit_logs, start=1):
            print(f"{i}. {log['type']}  {log['amount']:,}원  (처리 후 잔액: {log['balance_after']:,}원)")

    # (Q10) 출금 내역 출력
    def withdrawal_history(self):
        print("===== 출금 내역 =====")
        if len(self.withdraw_logs) == 0:
            print("내역이 없습니다.")
            return

        for i, log in enumerate(self.withdraw_logs, start=1):
            print(f"{i}. {log['type']}  {log['amount']:,}원  (처리 후 잔액: {log['balance_after']:,}원)")


# =========================
# (Q8) 3개 이상 인스턴스 생성 + 리스트 저장
# =========================
acc1 = Account_KS("홍길동", 1000000)
acc2 = Account_KS("이순신", 500000)
acc3 = Account_KS("안중근", 2000000)

accounts = [acc1, acc2, acc3]

# (Q3) 계좌 개수 출력
Account_KS.get_account_num()

print("\n[전체 계좌 정보]")
for acc in accounts:
    acc.display_info()

# =========================
# (Q9) 잔액 100만원 이상 고객만 출력
# =========================
print("\n[잔액 100만원 이상 고객]")
for acc in accounts:
    if acc.balance >= 1000000:
        acc.display_info()

# =========================
# (Q4, Q5, Q7, Q10) 입출금/이자 테스트
# =========================
print("\n[입출금 테스트 - acc2]")
acc2.deposit(1000)
acc2.deposit(2000)
acc2.deposit(3000)
acc2.deposit(4000)
acc2.deposit(5000)  # 5번째 입금 -> 이자 발생

acc2.withdraw(2000)
acc2.withdraw(999999999)  # 잔액 부족 테스트

print("\n[acc2 내역 출력]")
acc2.deposit_history()
acc2.withdrawal_history()

    
```
